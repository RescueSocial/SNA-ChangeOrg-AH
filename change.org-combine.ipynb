## Combine data from multiple .csv files

```python
import pandas as pd
import os
import re
from tqdm.auto import tqdm
from collections import Counter
```

```python
# filename pattern from get_last_signatures_csv.py
pattern = f"^change_org_petition_slug_(.+)" + \
                   f"_at_(.+)"+ \
                   f"_max_number_signatures_(.+).csv$"

# directory with outputs
data_dir = './data'

# format: petition slug -> list of dataframes
data_by_petition = {}

# loading all data
for f in tqdm(os.listdir(data_dir)):
    if re.match(pattern, f):
        data = re.search(pattern, f)
        df = pd.read_csv(os.path.join(data_dir, f))
        # data_by_petition/
        slug, datetime, max_signatures = data.groups()
        if slug not in data_by_petition:
            data_by_petition[slug] = []
            
        data_by_petition[slug].append(df)
```

```python
Counter([key for key, lst in data_by_petition.items() for item in lst])
```

```python
slug = 'hollywood-boycott-aquaman-2-amber-heard'
```

```python
df = pd.concat(data_by_petition[slug])
df = df.drop_duplicates()
df = df.reset_index()
df = df.sort_values('timestamp')
```
