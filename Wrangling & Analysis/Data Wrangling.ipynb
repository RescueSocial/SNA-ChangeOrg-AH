### Drop Columns

```python
drop_list = ['children_count', 'user.slug', 'user.first_name', 'user.last_name', 
             'user.short_display_name', 'commentable_entity.slug']
```

```python
df_comments.drop(columns=drop_list, inplace=True)
```

### Changing Columns Names

```python
df_comments.rename(columns={'user.id': 'user_id',
                            'user.display_name': 'username',
                            'user.city': 'user_city', 
                            'user.country_code': 'user_country_code', 
                            'user.locale': 'user_locale',  
                            'petition__slug': 'petition_slug'}, inplace=True)
```

```python
print(df_comments.shape)
df_comments.head(1)
```

### Changing Data Types

```python
df_comments.info()
```

```python
df_comments['id'] = df_comments['id'].astype(str)
df_comments['commentable_id'] = df_comments['commentable_id'].astype(str)
df_comments["created_at"] = pd.to_datetime(df_comments["created_at"])
df_comments['user_id'] = df_comments['user_id'].astype(str)
```

```python
df_comments['year'] = df_comments['created_at'].dt.year
df_comments['date'] = df_comments['created_at'].dt.date
df_comments['month'] = df_comments['created_at'].dt.strftime('%b')
df_comments['dayofmonth'] = df_comments['created_at'].dt.day
df_comments['hour'] = df_comments['created_at'].dt.hour
```

### Normalizing Text

```python
df_comments.comment = df_comments.comment.str.lower()
```

```python
print(df_comments.shape)
df_comments.head(1)
```

```python
df_comments.to_csv('./cleaned_data/petitions_comments_clened.csv', index=False)
```
