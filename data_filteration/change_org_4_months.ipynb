```python
def months_filteration(df, datetime_column, start, end):
    """
    Filter dataframe due to time, get data between two dates
    """
    return (
        df[
            (df[datetime_column].astype(str) >= start)
            & (df[datetime_column].astype(str) < end)
        ]
        .reset_index()
        .drop(columns="index")
    )
```
